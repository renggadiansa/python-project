```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler

dataset_x = pd.read_csv("trade1.csv")
dataset.head()
#melihat infor setiap variabel
dataset.info()

print("\n")
#menghilangkan kolom
dataset_x = dataset_x.drop(["Direction","Weekday","Measure","Country","Commodity","Transport_Mode","Date"], axis=1)
dataset_x.head()

#menentukan variabel yang akan di clusterkan
dataset_x = dataset_x.iloc[:, 1:3]
dataset_x.head()
print("\n")

#mengubah datamenjadi array
x_array =np.array(dataset_x)
print(x_array)

#mengembalikan ukuran variabel menjad standar
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x_array)
x_scaled

kmeans = KMeans(n_clusters= 10, random_state=123 )
kmeans.fit(x_scaled)

#menampilkan hasil kluster
print(kmeans.labels_)
#menambahkan kolom "kluster" dalam data frame dataset
dataset_x["kluster"] = kmeans.labels_
dataset_x.head(20)
```
