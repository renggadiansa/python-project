```python
import pandas as pd

# Membuat dataframe contoh dengan nilai null
# df = pd.DataFrame({'A': [1, 2, None, 4],
#                    'B': [5, None, 7, 8],
#                    'C': [9, 10, 11, 12]})

df = pd.read_csv('../ai/BankChurners.csv')

# Menampilkan dataframe sebelum membersihkan nilai null
print("Dataframe sebelum membersihkan nilai null:\n", df)

# Membersihkan nilai null dengan fungsi dropna()
df_clean = df.dropna()

# Menampilkan dataframe setelah membersihkan nilai null
print("\nDataframe setelah membersihkan nilai null:\n", df_clean)
```
