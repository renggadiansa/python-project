```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
```

```python
df = pd.read_csv('jaringan.csv')
```

```python
# Membangun objek graf
G = nx.from_pandas_edgelist(df, 'source', 'target')

# Menentukan posisi simpul menggunakan algoritma spring layout
pos = nx.spring_layout(G)

# Menggambar simpul dan sisi graf
nx.draw_networkx_nodes(G, pos, node_color='lightblue', node_size=500)
nx.draw_networkx_edges(G, pos, edge_color='gray')

# Menampilkan label pada simpul
labels = {node: node for node in G.nodes()}
nx.draw_networkx_labels(G, pos, labels, font_size=10, font_color='black')

# Menampilkan graf
plt.axis('off')
plt.show()
```

```python
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=500, edge_color='gray', width=0.5)
plt.show()
```

```python
# Membuat objek graf
G = nx.Graph()

# Menambahkan simpul (node) ke graf
G.add_nodes_from([1, 2, 3, 4, 5])

# Menambahkan sisi (edge) ke graf
G.add_edges_from([(1, 2), (1, 3), (2, 3), (3, 4), (4, 5)])

# Menentukan posisi simpul menggunakan algoritma spring layout
pos = nx.spring_layout(G)

# Menggambar simpul dan sisi graf
nx.draw_networkx_nodes(G, pos, node_color='lightblue', node_size=500)
nx.draw_networkx_edges(G, pos, edge_color='gray')

# Menambahkan label pada simpul
labels = {1: 'John Doe', 2: 'Jane Smith', 3: 'Michael Johnson', 4: 'Emily Brown', 5: 'David Wilson'}
nx.draw_networkx_labels(G, pos, labels, font_size=10, font_color='black')

# Menampilkan graf
plt.axis('off')
plt.show()
```
