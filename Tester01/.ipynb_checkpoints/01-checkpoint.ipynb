# DATA PREPARATION

```python
# Combine Data
data1 = pd.read_csv('employ.csv')
data2 = pd.read_csv('finance.csv')

frames = [data1, data2]
results = pd.concat(frames).drop_duplicates().reset_index(drop = True)

# Formating Data
results.to_csv(r'new_value.csv', index = False)
```

```python
X = df2.iloc[:, [0,1]].values

from sklearn.cluster import KMeans
wcss=[]
for i in range (1,10):
    kmeans = KMeans(n_clusters = i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,10),wcss)
plt.show()
```
